There are no missing values and all the variables are of appropriate datatype.

However, there may be situations where there may be missing values present in the dataset. In that case, they need to be imputed with the mean or the median value, whichever is suitable. The datatypes also need to be converted to the appropriate ones.

```
data.describe()
```

It is to be noted that there are columns whose range (minimum, maximum) vary a lot. For eg., TARGET_deathRate lies between [0, 1] and BirthRate lies between [0, 21.32].

So, we need to scale all the variables to the same range.



```
data_ = (data - data.mean())/data.std()
data_.head()
```

```
data_.describe()
```

```
int(-8.732030e-15)
```
